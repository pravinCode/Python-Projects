```python
import pandas as pd
confirmed_cases = pd.read_csv("covid_19_confirmed.csv")
confirmed_cases
```

```python
death_cases = pd.read_csv("covid_19_deaths.csv")
death_cases
```

```python
death_cases.columns = death_cases.iloc[0]
death_cases = death_cases.drop(0).reset_index(drop = True)
death_cases
```

```python
recovered_cases = pd.read_csv("covid_19_recovered.csv")
recovered_cases
```

```python
recovered_cases.columns = recovered_cases.iloc[0]
recovered_cases = recovered_cases.drop(0).reset_index(drop = True)
recovered_cases
```

```python
id_vars = ['Province/State', 'Country/Region', 'Lat', 'Long']
all_cols = confirmed_cases.columns.tolist()
date_cols = [c for c in all_cols if c not in id_vars]
date_cols
```

```python
# confirmed [ Melting - unpivot]
confirmed_long = confirmed_cases.melt(
    id_vars = id_vars,
    value_vars = date_cols,
    var_name = "Date",
    value_name = "Confirmed"
)
confirmed_long
```

```python
# Deaths [Melting - unpivot]
id_vars = ['Province/State', 'Country/Region', 'Lat', 'Long']
all_cols = death_cases.columns.tolist()
date_cols = [c for c in all_cols if c not in id_vars]
deaths_long = death_cases.melt(
    id_vars = id_vars,
    value_vars = date_cols,
    var_name = "Date",
    value_name = "Deaths"
)
deaths_long
```

```python
# Recovered [Melting - unpivot]
id_vars = ['Province/State', 'Country/Region', 'Lat', 'Long']
all_cols = recovered_cases.columns.tolist()
date_cols = [c for c in all_cols if c not in id_vars]
recovered_long = recovered_cases.melt(
    id_vars = id_vars,
    value_vars = date_cols,
    var_name = "Date",
    value_name = "Recovered"
)
recovered_long
```

```python
# 5.1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

target_countries = ['Germany', 'France', 'Italy']
country_date_total = (
    merged.loc[merged['Country/Region'].isin(target_countries)]
    .groupby(['Country/Region','Date'])['Confirmed']
    .sum().reset_index()
    .sort_values(['Country/Region','Date'])
)
country_date_total
```

```python
country_date_total['DailyNew'] = (
    country_date_total
    .groupby('Country/Region')['Confirmed']
    .diff()
    .fillna(country_date_total['Confirmed'])
)
country_date_total
```

```python
daily_new_matrix = (
    country_date_total.pivot(
        index="Date",
        columns="Country/Region",
        values="DailyNew"
    ).sort_index()
)

daily_new_matrix
```

```python
daily_new_by_country = {}
for country in target_countries:
    if country in daily_new_matrix.columns:
        s= daily_new_matrix[country]
        daily_new_by_country[country] = s

daily_new_by_country
```

```python
daily_new_matrix[country]
```

```python
# Peaks[Country] : (max_value, date per date)
peaks = {}
for country, series in daily_new_by_country.items():
    max_value = int(series.max())
    max_index = series.idxmax()
    peaks[country] = (max_value, max_index)

peaks
```

```python
print("Peak Daily New Cases (value, date):")
for country, (values, when) in peaks.items():
    print(f"{country} -> {values} on {when.date()}")
```

```python
plt.figure(figsize = (12,6))
for country, series in daily_new_by_country.items():
    plt.plot(series.index, series.values, label = country)

plt.title("Daily New COVID-19 cases - Germany, France, Italy", color="red")
plt.xlabel("Date")
plt.ylabel("New Cases")
plt.xticks(rotation=45)
plt.ylim(-50000,200000)
plt.legend()
plt.show()
```

```python
#Q-5.3 What is the distribution of death rates (deaths/confirmed cases) among provinces in canada?
# Identify the province with the highest and lowest death rate as of the latest data point.

canada_confirmed_cases = (
    confirmed_long[confirmed_long['Country/Region'] == 'Canada']
    .sort_values('Date')
    .groupby(['Province/State']).tail(1)[['Province/State','Confirmed']]
    .set_index('Province/State')
)

canada_confirmed_cases
```

```python
canada_death_cases = (
    deaths_long[deaths_long['Country/Region'] == 'Canada']
    .sort_values('Date')
    .groupby(['Province/State']).tail(1)[['Province/State','Deaths']]
    .set_index('Province/State')
)

canada_death_cases
```

```python
import numpy as np
canada = canada_confirmed_cases.join(canada_death_cases, how = "inner")
canada['DeathRate'] = (canada['Deaths'] / canada['Confirmed'] * 100).round(2).replace(np.inf,0)
canada
```

```python
import matplotlib.pyplot as plt

# Filter only provinces with DeathRate > 0
df_nonzero = canada[canada['DeathRate'] > 0].sort_values('DeathRate', ascending=False)

plt.figure(figsize=(10, 8))  # smaller figure size

bars = plt.bar(
    df_nonzero.index,           
    df_nonzero['DeathRate'],    
    color= ["red","green","blue","brown","lightblue","lightgreen","yellow","orange","indigo","violet","green"]
)

# Add bar labels
plt.bar_label(
    bars,
    labels=[f"{b.get_height():.2f}%" for b in bars],
    padding=5
)

# X-axis labels
plt.xticks(rotation=45)

plt.title('COVID-19 Provinces Death Rate in Canada')
plt.xlabel('Province/State')
plt.ylabel('Death Rate %')

plt.tight_layout()
plt.show()
```

```python
m = merged.copy()
m = m.sort_values(['Country/Region','Date'])
m['YearMonth'] = m['Date'].dt.to_period('M')
m
```

```python
# Daily Confirmed
confirmed_diff = m.groupby('Country/Region')['Confirmed'].diff().clip(lower = 0)
m['DailyConfirmed'] = confirmed_diff.fillna(0)

# Daily Deaths
deaths_diff = m.groupby('Country/Region')['Deaths'].diff().clip(lower = 0)
m['DailyDeaths'] = deaths_diff.fillna(0)

# Daily Recovered
recovered_diff = m.groupby('Country/Region')['Recovered'].diff().clip(lower = 0)
m['DailyRecovered'] = recovered_diff.fillna(0)

m
```

```python
monthly_sum_trends = (
    m.groupby(['Country/Region', 'YearMonth'])[['DailyConfirmed','DailyDeaths','DailyRecovered']]
    .sum()
    .rename(
        columns = {
            'DailyConfirmed': 'Confirmed',
            'DailyDeaths':'Deaths',
            'DailyRecovered':'Recovered',
        }).astype(int)
    .reset_index()
    .sort_values(['Country/Region', 'YearMonth'])
)

monthly_sum_trends
```

```python
# Q-8.1 For the combined dataset, identify the three countries with the highest average death rates (deaths/confirmed cases) throughout 2020.
# what might this indicate about the pandemic's impact in these countries?

df = monthly_sum_trends.copy()
df = df.sort_values(['Country/Region','YearMonth'])
df_2020 = df[(df['YearMonth'] >= '2020-01') & (df['YearMonth'] <= '2020-12')]
df_2020
```

```python
# Aggregate totals[Confirmed, Death] over 200
totals_2020 =(
    df_2020.groupby('Country/Region')[['Confirmed','Deaths']]
    .sum()
    .rename(
        columns = {
            'Confirmed': 'Confirmed2020',
            'Deaths': 'Deaths2020'
        }
    )
)
totals_2020
```

```python
totals_2020['DeathRate'] = (totals_2020['Deaths2020']/totals_2020['Confirmed2020']*100).round(2)
totals_2020
```

```python
top_3 = totals_2020.sort_values('DeathRate', ascending = False).head(3)
top_3
```

```python
ax= top_3['DeathRate'].plot(kind='bar', figsize = (8,4), color=["Green", "Red", "Blue"])
ax.set_title("Top 3 Countries with Highest Average Death Rate (2020)", color = "Red")
ax.set_xlabel("Country")
ax.set_ylabel("Death Rate(%)")
ax.bar_label(ax.containers[0], fmt='%.2f%%')
plt.ylim(0,40)
plt.xticks(rotation=45)
plt.show()
```

```python
# Q-8.3 Analyze the ratio of recoveries to confirmed cases for the United States on monthly basis from March 2020 to May 2021.
# Which month experienced the highest recovery ratio, and what could be the potential reasons?

us_data = monthly_sum_trends.copy()
us_data = us_data[us_data['Country/Region'] == 'US'].sort_values('YearMonth').set_index('YearMonth')
us_data
```

```python
recovery_ratio_us = (
    us_data['Recovered'] / us_data['Confirmed']
)

recovery_ratio_us = recovery_ratio_us.loc[pd.Period('2020-03', freq='M'):pd.Period('2021-05', freq='M')]
recovery_ratio_us
```

```python
# plot the recovery_ratio_us using line plot

fig , ax = plt.subplots(figsize = (10,6))
recovery_ratio_us.plot(kind="line", marker="o",linestyle="-", ax=ax)
ax.set_title("Monthly Recovery Ratio in the United States")
ax.set_xlabel("Month")
ax.set_ylabel("Recovery Ratio")
ax.grid(True, alpha=0.3)
plt.show()
```
